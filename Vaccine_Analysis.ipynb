```python
vaccine_brand_data = statewise.groupby('State').agg({
    'Covaxin_DA': 'sum',
    'CoviShield_DA': 'sum',
    'Sputnik_DA': 'sum'
}).reset_index()
vaccine_brand_data['Total_Doses'] = vaccine_brand_data['Covaxin_DA'] + vaccine_brand_data['CoviShield_DA'] + vaccine_brand_data['Sputnik_DA']
vaccine_brand_data['Covaxin_Percent'] = (vaccine_brand_data['Covaxin_DA'] / vaccine_brand_data['Total_Doses']) * 100
vaccine_brand_data['CoviShield_Percent'] = (vaccine_brand_data['CoviShield_DA'] / vaccine_brand_data['Total_Doses']) * 100
vaccine_brand_data['Sputnik_Percent'] = (vaccine_brand_data['Sputnik_DA'] / vaccine_brand_data['Total_Doses']) * 100
fig, ax = plt.subplots(figsize=(12, 8))
vaccine_brand_data.plot(
    x='State',
    kind='barh',
    stacked=True,
    color=['#ff9999', '#66b3ff', '#99ff99'],
    ax=ax
)
ax.set_xlabel('Total Doses Administered (%)')
ax.set_ylabel('State')
ax.set_title('Vaccine Brand Distribution by State')
plt.show()
```

```python
data_timeseries = pd.read_csv('covid_vaccine_statewise.csv')
columns=["Update","State","TDA","Sessions","Sites","FDA","SDA","Male_DA","Female_DA","Transgender_DA","Covaxin_DA","CoviShield_DA","Sputnik_DA","AEFI","18-44_DA","45-60_DA","60+DA","18-44_IV","45-60_IV","60+IV","Male_IV","Female_IV","Transgender_IV","TIV"]
data_timeseries.columns=columns
data_timeseries.head()
```

```python
data_timeseries['Update'] = pd.to_datetime(data_timeseries['Update'])
data_timeseries.set_index('Update', inplace=True)
monthly_grouped = data_timeseries.groupby('State').resample('M').agg({
    'TDA': 'sum',  # Total Doses Administered
    'FDA': 'sum',  # First Dose Administered
    'SDA': 'sum'   # Second Dose Administered
}).reset_index()
monthly_grouped['Cumulative_TDA'] = monthly_grouped.groupby('State')['TDA'].cumsum()
monthly_grouped['Cumulative_FDA'] = monthly_grouped.groupby('State')['FDA'].cumsum()
monthly_grouped['Cumulative_SDA'] = monthly_grouped.groupby('State')['SDA'].cumsum()
top_states = monthly_grouped.groupby('State')['Cumulative_TDA'].max().nlargest(5).index
fig, axes = plt.subplots(3, 1, figsize=(12, 12))
for state in top_states:
    state_data = monthly_grouped[monthly_grouped['State'] == state]
    axes[0].plot(state_data['Update'], state_data['Cumulative_TDA'], label=state)
axes[0].set_title('Cumulative Total Doses Administered')
axes[0].set_xlabel('Date')
axes[0].set_ylabel('Cumulative Total Doses')
axes[0].legend()
for state in top_states:
    state_data = monthly_grouped[monthly_grouped['State'] == state]
    axes[1].plot(state_data['Update'], state_data['Cumulative_FDA'], label=state)
axes[1].set_title('Cumulative First Doses Administered')
axes[1].set_xlabel('Date')
axes[1].set_ylabel('Cumulative First Doses')
axes[1].legend()
for state in top_states:
    state_data = monthly_grouped[monthly_grouped['State'] == state]
    axes[2].plot(state_data['Update'], state_data['Cumulative_SDA'], label=state)
axes[2].set_title('Cumulative Second Doses Administered')
axes[2].set_xlabel('Date')
axes[2].set_ylabel('Cumulative Second Doses')
axes[2].legend()
plt.tight_layout()
plt.savefig('timeseries.png')
plt.show()
```

```python
global_data=pd.read_csv('global_data.csv')
global_data.head()
```

```python
global_data.columns
```

```python
global_data['testing_efficiency'] = global_data['total_confirmed'] / global_data['total_tests']
global_data = global_data.dropna(subset=['testing_efficiency'])
global_data=global_data.sort_values(by='testing_efficiency',ascending=False)
top_efficiency=global_data.head(15)
plt.figure(figsize=(12,6))
sns.barplot(x='country', y='testing_efficiency', data=top_efficiency,errorbar=None,linewidth=2,color="cyan",edgecolor="black")
plt.title('Testing Efficiency by Country (Total Confirmed Cases / Total Tests)')
plt.xlabel('Country')
plt.ylabel('Testing Efficiency')
plt.xticks(rotation=90)
plt.show()
```

```python
continent_testing_efficiency = global_data.groupby('continent')['testing_efficiency'].mean()
plt.figure(figsize=(10, 6))
sns.barplot(x=continent_testing_efficiency.index, y=continent_testing_efficiency.values, errorbar=None,linewidth=2,color="lightgreen",edgecolor="black")
plt.title('Mean Testing Efficiency by Continent')
plt.xlabel('Continent')
plt.ylabel('Mean Testing Efficiency')
plt.show()
```

```python
correlation_matrix = global_data[['total_confirmed', 'total_deaths', 'total_recovered', 'active_cases', 'serious_or_critical', 'total_tests']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
